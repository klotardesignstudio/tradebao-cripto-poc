```python
import requests
import pandas as pd
from datetime import datetime

# 1. Parmetros da Requisio
base_url = "https://data-api.binance.vision"
endpoint = "/api/v3/klines"
params = {
    "symbol": "LINKBTC",
    "interval": "1d",
    "limit": 1000 # Pegando os ltimos 1000 dias de dados
}

# Define o nmero de linhas a serem exibidas
num_linhas_exibir = 1000  # Voc pode alterar este valor para ver mais ou menos linhas

# 2. Fazendo a Requisio
try:
    response = requests.get(base_url + endpoint, params=params)
    response.raise_for_status() # Lana um erro se a requisio falhar
    data = response.json()

    # 3. Processando e Estruturando os Dados com Pandas
    # A API retorna uma lista de listas. Vamos definir os nomes das colunas.
    columns = [
        "Open Time", "Open", "High", "Low", "Close", "Volume",
        "Close Time", "Quote Asset Volume", "Number of Trades",
        "Taker Buy Base Asset Volume", "Taker Buy Quote Asset Volume", "Ignore"
    ]

    df = pd.DataFrame(data, columns=columns)

    # 4. Limpeza e Converso de Tipos de Dados
    # Convertendo as colunas de preo e volume para nmeros (float)
    numeric_cols = ["Open", "High", "Low", "Close", "Volume", "Quote Asset Volume"]
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col])

    # Convertendo o tempo (que est em milissegundos) para um formato legvel
    df["Open Time"] = pd.to_datetime(df["Open Time"], unit='ms')
    df["Close Time"] = pd.to_datetime(df["Close Time"], unit='ms')

    # Selecionando apenas as colunas que nos interessam
    df_final = df[["Open Time", "Open", "High", "Low", "Close", "Volume"]]

    # Usando a data de abertura como nosso ndice para facilitar a anlise
    df_final.set_index("Open Time", inplace=True)


    print("Dados capturados com sucesso!")

    # Configura o nmero mximo de linhas a serem exibidas
    pd.set_option('display.max_rows', num_linhas_exibir)

    display(df_final)

except requests.exceptions.RequestException as e:
    print(f"Erro ao fazer a requisio: {e}")
```
